```python
import os
with open("loocv_result.txt", 'r') as f:
    data = f.readlines()
```

```python
for i in range(12):
    acc, auc, pre, recall = [], [], [], []
    for j in range(5):
        info = data[i + 14 * j].split()
        acc.append(float(info[1][4: -1]))
        auc.append(float(info[2][4: -1]))
        pre.append(float(info[-2][10: -1]))
        recall.append(float(info[-1][7: -1]))
        print(info)

    print("acc: ", round(sum(acc) / 5, 4), "auc: ", round(sum(auc) / 5, 4),
          "pre: ", round(sum(pre) / 5, 4), "recall: ", round(sum(recall) / 5, 4))
    print("\n\n\n")
```
